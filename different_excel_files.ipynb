3. .xls files use the binary file format, which is less efficient than the newer .xlsx format, which uses the Open XML file format. Open XML is a newer, XML-based format that is more compact and can better represent complex data structures with less redundancy. the .xlsx format also uses techniques such as ZIP compression to store the data more efficiently. Hence, .xls files are much larger than .xlsx files.
